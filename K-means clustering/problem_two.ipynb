```
# Refers to scikit-image, a Python library for image processing
# Submodule & handles input and output operations for images
from skimage import io
# Scikit-learn library - provides tools for machine learning and data mining.
# Grouping data into clusters
from sklearn.cluster import KMeans
# Support for large, multi-dimensional arrays and matrices
# Shorthand for numpy
import numpy as np
# Creating static, animated, and interactive visualizations in Python
# submodule - creating plots and graphs, used to shorten function calls
import matplotlib.pyplot as plt
```

```
# Image Load
image = io.imread('/content/converted.png')

# Check image alpha channel
if image.shape[2] == 4:
    image = image[:, :, :3]  # Discard alpha channel

io.imshow(image)
plt.title("Original Image")
plt.show()
```

```
# Image reshape into a 2D array of pixels
rows, cols, ch = image.shape  # height, width, channels
image_flat = image.reshape(rows * cols, 3)  # total number of pixels; RGB values
print(f"Image shape: {image.shape}") # prints original shape of the image before reshaping
print(f"Flattened image shape: {image_flat.shape}") # prints shape after reshaping
```

```
# Set k value
k = 8

# Apply k-Means clustering
#This parameter specifies the number of clusters
kmeans = KMeans(n_clusters=k, n_init=10)  # This parameter defines how many times the K-Means algorithm will be run
kmeans.fit(image_flat) # flattened image data

# Replace each pixel with its corresponding cluster center
labels = kmeans.labels_   # labels_ represents which cluster out of k clusters
# After the K-Means algorithm has finished; centers represent the average color for each
centers = np.asarray(kmeans.cluster_centers_, dtype=np.uint8)

# Reconstruct the compressed image
# Maps each pixel to its corresponding cluster center’s RGB values
compressed_image = centers[labels].reshape(rows, cols, 3) # pixel colors back into the original image dimensions

# Display the compressed image
plt.imshow(compressed_image)
plt.title(f'Compressed Image with K = {k}')
plt.show()
```
