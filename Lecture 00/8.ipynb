```python
sentence = "welcome to ai lab"
freq = {}
for character in sentence:
  if character not in freq.keys():
    freq[character] = 0
  freq[character] += 1
print(freq)
```

```python
import math
def myFunction(x1, y1, x2, y2):
  a = x1 - y1
  b = x2 - y2
  dist = math.sqrt(a**2 + b**2)
  return dist

myFunction(0, 3, 4, 0)
```
