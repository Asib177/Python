```python
grid = [
    ".....#",
    ".....#",
    "..#..#",
    "E..#..",
    "..#...",
    ".#G..."
]

cost = {}
vis = {}

def bfs(startpoint):
    cost[startpoint] = 0
    vis[startpoint] = True
    q = [startpoint]

    while q:
        u = q[0]
        q.pop(0)

        row = u[0]
        col = u[1]

        # Check all four possible directions: up, down, left, right

        # Up
        if row - 1 >= 0 and grid[row - 1][col] != "#":
            if (row - 1, col) not in vis:
                cost[(row - 1, col)] = cost[u] + 1
                vis[(row - 1, col)] = True
                q.append((row - 1, col))

        # Down
        if row + 1 < len(grid) and grid[row + 1][col] != "#":
            if (row + 1, col) not in vis:
                cost[(row + 1, col)] = cost[u] + 1
                vis[(row + 1, col)] = True
                q.append((row + 1, col))

        # Left
        if col - 1 >= 0 and grid[row][col - 1] != "#":
            if (row, col - 1) not in vis:
                cost[(row, col - 1)] = cost[u] + 1
                vis[(row, col - 1)] = True
                q.append((row, col - 1))

        # Right
        if col + 1 < len(grid[0]) and grid[row][col + 1] != "#":
            if (row, col + 1) not in vis:
                cost[(row, col + 1)] = cost[u] + 1
                vis[(row, col + 1)] = True
                q.append((row, col + 1))

    print(cost)

bfs((3, 0))
```
